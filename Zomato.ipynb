```python
#  Clean Users Data

print("="*60)
print("CLEANING USERS DATA")
print("="*60)

# Check original shape
print(f"Original shape: {user_df.shape}")

# 1. Rename column if needed (ensure consistency)
user_df.rename(columns={'id': 'user_id'}, inplace=True, errors='ignore')

# 2. Remove duplicates
user_df = user_df.drop_duplicates(subset=['user_id'], keep='first')
print(f"After removing duplicates: {user_df.shape}")

# 3. Clean text columns
text_columns = ['name', 'gender', 'marital_status', 'occupation']
for col in text_columns:
    if col in user_df.columns:
        user_df[col] = user_df[col].astype(str).str.strip().str.upper()
        user_df[col] = user_df[col].replace(['NAN', 'NONE', ''], np.nan)

# 4. Standardize gender
if 'gender' in user_df.columns:
    user_df['gender'] = user_df['gender'].map({
        'M': 'MALE', 'MALE': 'MALE', 'M': 'MALE',
        'F': 'FEMALE', 'FEMALE': 'FEMALE',
        'O': 'OTHER', 'OTHER': 'OTHER'
    }).fillna('UNKNOWN')

# 5. Standardize marital status
if 'marital_status' in user_df.columns:
    user_df['marital_status'] = user_df['marital_status'].map({
        'S': 'SINGLE', 'SINGLE': 'SINGLE',
        'M': 'MARRIED', 'MARRIED': 'MARRIED',
        'D': 'DIVORCED', 'DIVORCED': 'DIVORCED',
        'W': 'WIDOWED', 'WIDOWED': 'WIDOWED'
    }).fillna('UNKNOWN')

# 6. Handle age outliers
if 'age' in user_df.columns:
    user_df.loc[user_df['age'] < 18, 'age'] = np.nan
    user_df.loc[user_df['age'] > 100, 'age'] = np.nan
    # Fill missing ages with median
    user_df['age'].fillna(user_df['age'].median(), inplace=True)

# 7. Create age groups
if 'age' in user_df.columns:
    user_df['age_group'] = pd.cut(user_df['age'], 
                                 bins=[0, 18, 25, 35, 45, 60, 100],
                                 labels=['<18', '18-25', '26-35', '36-45', '46-60', '60+'])

print(f"Final shape: {user_df.shape}")
print("\n✅ Users data cleaned successfully!")
print("\nSample cleaned data:")
print(user_df.head())
```

```python
#  Clean Restaurant Data

print("="*60)
print("CLEANING RESTAURANT DATA")
print("="*60)

print(f"Original shape: {resturant_df.shape}")

# 1. Rename columns for consistency
resturant_df.rename(columns={'id': 'restaurant_id'}, inplace=True, errors='ignore')

# 2. Remove duplicates
resturant_df = resturant_df.drop_duplicates(subset=['restaurant_id'], keep='first')
print(f"After removing duplicates: {resturant_df.shape}")

# 3. Clean text columns
text_cols = ['name', 'country', 'city', 'cuisine']
for col in text_cols:
    if col in resturant_df.columns:
        resturant_df[col] = resturant_df[col].astype(str).str.strip().str.upper()
        resturant_df[col] = resturant_df[col].replace(['NAN', 'NONE', ''], np.nan)

# 4. Handle rating (convert to numeric) - FIXED VERSION
if 'rating' in resturant_df.columns:
    resturant_df['rating'] = pd.to_numeric(resturant_df['rating'], errors='coerce')
    # Fill missing ratings with median - removed inplace=True
    resturant_df['rating'] = resturant_df['rating'].fillna(resturant_df['rating'].median())
    # Round to 1 decimal
    resturant_df['rating'] = resturant_df['rating'].round(1)

# 5. Handle rating_count - FIXED VERSION (if you have similar issue)
if 'rating_count' in resturant_df.columns:
    resturant_df['rating_count'] = pd.to_numeric(resturant_df['rating_count'], errors='coerce')
    resturant_df['rating_count'] = resturant_df['rating_count'].fillna(0)

# 6. Create rating category
if 'rating' in resturant_df.columns:
    resturant_df['rating_category'] = pd.cut(resturant_df['rating'],
                                            bins=[0, 2.5, 3.5, 4.0, 4.5, 5.0],
                                            labels=['Poor', 'Average', 'Good', 'Very Good', 'Excellent'])

print(f"Final shape: {resturant_df.shape}")
print("\n✅ Restaurant data cleaned successfully!")
print("\nSample cleaned data:")
print(resturant_df.head())
```

```python
#  Clean Orders Data

from datetime import datetime
print("="*60)
print("CLEANING ORDERS DATA")
print("="*60)

print(f"Original shape: {order_df.shape}")

# 1. Rename columns
order_df.rename(columns={'r_id': 'restaurant_id'}, inplace=True, errors='ignore')

# 2. Convert order_date to datetime
if 'order_date' in order_df.columns:
    order_df['order_date'] = pd.to_datetime(order_df['order_date'], errors='coerce')
    # Remove future dates
    order_df = order_df[order_df['order_date'] <= datetime.now()]
    print(f"After removing future dates: {order_df.shape}")

# 3. Handle sales_quantity
if 'sales_quantity' in order_df.columns:
    order_df['sales_quantity'] = pd.to_numeric(order_df['sales_quantity'], errors='coerce')
    order_df.loc[order_df['sales_quantity'] <= 0, 'sales_quantity'] = 1
    order_df['sales_quantity'].fillna(1, inplace=True)

# 4. Handle sales_amount
if 'sales_amount' in order_df.columns:
    order_df['sales_amount'] = pd.to_numeric(order_df['sales_amount'], errors='coerce')
    order_df.loc[order_df['sales_amount'] <= 0, 'sales_amount'] = np.nan

# 5. Calculate revenue
if 'sales_quantity' in order_df.columns and 'sales_amount' in order_df.columns:
    order_df['revenue'] = order_df['sales_quantity'] * order_df['sales_amount']
    # Fill missing revenue with median
    order_df['revenue'].fillna(order_df['revenue'].median(), inplace=True)

# 6. Extract time features
if 'order_date' in order_df.columns:
    order_df['order_month'] = order_df['order_date'].dt.strftime('%Y-%m')
    order_df['order_day'] = order_df['order_date'].dt.day
    order_df['order_day_of_week'] = order_df['order_date'].dt.day_name()
    order_df['order_hour'] = order_df['order_date'].dt.hour
    order_df['order_weekend'] = order_df['order_day_of_week'].isin(['Saturday', 'Sunday'])

# 7. Remove outliers (top 1% revenue)
if 'revenue' in order_df.columns:
    revenue_99th = order_df['revenue'].quantile(0.99)
    orders_before = len(order_df)
    order_df = order_df[order_df['revenue'] <= revenue_99th]
    print(f"After removing revenue outliers: {order_df.shape} (removed {orders_before - len(order_df)} rows)")

print(f"Final shape: {order_df.shape}")
print("\n✅ Orders data cleaned successfully!")
print("\nSample cleaned data:")
print(order_df.head())
```

```python
#  Clean Menu Data

print("="*60)
print("CLEANING MENU DATA")
print("="*60)

print(f"Original shape: {menu_df.shape}")

# 1. Rename columns
menu_df.rename(columns={'r_id': 'restaurant_id', 'f_id': 'food_id'}, inplace=True, errors='ignore')

# 2. Remove duplicates
menu_df = menu_df.drop_duplicates(subset=['menu_id'], keep='first')
print(f"After removing duplicates: {menu_df.shape}")

# 3. Clean text columns
if 'cuisine' in menu_df.columns:
    menu_df['cuisine'] = menu_df['cuisine'].astype(str).str.strip().str.upper()
    menu_df['cuisine'] = menu_df['cuisine'].replace(['NAN', 'NONE', ''], np.nan)

# 4. Handle price
if 'price' in menu_df.columns:
    menu_df['price'] = pd.to_numeric(menu_df['price'], errors='coerce')
    menu_df.loc[menu_df['price'] <= 0, 'price'] = np.nan
    # Fill missing prices with median by cuisine
    menu_df['price'] = menu_df.groupby('cuisine')['price'].transform(lambda x: x.fillna(x.median()))
    # Fill remaining NaN values with overall median (without inplace=True)
    menu_df['price'] = menu_df['price'].fillna(menu_df['price'].median())

print(f"Final shape: {menu_df.shape}")
print("\n✅ Menu data cleaned successfully!")
print("\nSample cleaned data:")
print(menu_df.head())
```

```python
#  Clean Food Data

print("="*60)
print("CLEANING FOOD DATA")
print("="*60)

print(f"Original shape: {food_df.shape}")

# 1. Rename columns
food_df.rename(columns={'f_id': 'food_id'}, inplace=True, errors='ignore')

# 2. Clean food item names
if 'item' in food_df.columns:
    food_df['item'] = food_df['item'].astype(str).str.strip().str.title()

# 3. Standardize veg_or_non_veg
if 'veg_or_non_veg' in food_df.columns:
    food_df['veg_or_non_veg'] = food_df['veg_or_non_veg'].astype(str).str.strip().str.upper()
    food_df['veg_or_non_veg'] = food_df['veg_or_non_veg'].map({
        'VEG': 'VEG', 'VEGETARIAN': 'VEG', 'V': 'VEG',
        'NON-VEG': 'NON-VEG', 'NONVEG': 'NON-VEG', 'NV': 'NON-VEG'
    }).fillna('UNKNOWN')

print(f"Final shape: {food_df.shape}")
print("\n✅ Food data cleaned successfully!")
print("\nSample cleaned data:")
print(food_df.head())
```

```python
# Clean Order Type Data

print("="*60)
print("CLEANING ORDER TYPE DATA")
print("="*60)

print(f"Original shape: {order_type_df.shape}")

# 1. Clean order_type column
if 'order_type' in order_type_df.columns:
    order_type_df['order_type'] = order_type_df['order_type'].astype(str).str.strip().str.upper()
    order_type_df['order_type'] = order_type_df['order_type'].map({
        'VEG': 'VEG', 'VEGETARIAN': 'VEG',
        'NON-VEG': 'NON-VEG', 'NONVEG': 'NON-VEG'
    }).fillna('UNKNOWN')

print("\n✅ Order type data cleaned successfully!")
print("\nSample cleaned data:")
print(order_type_df.head())
```

```python
#  Create Master Dataset

print("="*60)
print("CREATING MASTER DATASET")
print("="*60)

# First, let's check what column names we have
print("Checking column names...")
print(f"order_df columns: {order_df.columns.tolist()}")
print(f"order_type_df columns: {order_type_df.columns.tolist()}")

# Start with orders
master_df = order_df.copy()
print(f"\nStep 1 - Orders: {master_df.shape}")

# Merge with users
master_df = master_df.merge(user_df, on='user_id', how='left')
print(f"Step 2 - After adding users: {master_df.shape}")

# Merge with restaurant
master_df = master_df.merge(resturant_df, on='restaurant_id', how='left')
print(f"Step 3 - After adding restaurants: {master_df.shape}")

# Merge with menu
master_df = master_df.merge(menu_df[['restaurant_id', 'food_id', 'cuisine', 'price']], 
                            on=['restaurant_id'], how='left')
print(f"Step 4 - After adding menu: {master_df.shape}")

# Merge with food
master_df = master_df.merge(food_df, on='food_id', how='left')
print(f"Step 5 - After adding food: {master_df.shape}")

# Find the correct order ID column name
order_id_col = None
if 'order_id' in master_df.columns:
    order_id_col = 'order_id'
elif 'Order_Id' in master_df.columns:
    order_id_col = 'Order_Id'
elif 'orderid' in master_df.columns:
    order_id_col = 'orderid'
elif 'OrderID' in master_df.columns:
    order_id_col = 'OrderID'

# Find the correct column name in order_type_df
order_type_col = None
if 'order_id' in order_type_df.columns:
    order_type_col = 'order_id'
elif 'Order_Id' in order_type_df.columns:
    order_type_col = 'Order_Id'
elif 'orderid' in order_type_df.columns:
    order_type_col = 'orderid'
elif 'OrderID' in order_type_df.columns:
    order_type_col = 'OrderID'

print(f"\nMerging on: {order_id_col} -> {order_type_col}")

# Merge with order_type
if order_id_col and order_type_col:
    master_df = master_df.merge(order_type_df, left_on=order_id_col, right_on=order_type_col, how='left')
    print(f"Step 6 - After adding order type: {master_df.shape}")
else:
    print("ERROR: Could not find order ID column!")
    print(f"master_df columns: {master_df.columns.tolist()}")
    print(f"order_type_df columns: {order_type_df.columns.tolist()}")

print(f"\n✅ Final Master Dataset: {master_df.shape[0]:,} rows × {master_df.shape[1]} columns")
print(f"\nColumns in master dataset:")
for col in master_df.columns:
    print(f"  • {col}")
```

```python
#  Customer Demographics Visualization
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

fig = plt.figure(figsize=(20, 12))

# Auto-detect column names function
def find_column(df, *possible_names):
    for name in possible_names:
        if name in df.columns:
            return name
        for col in df.columns:
            if col.lower() == name.lower():
                return col
    return None

# Find all columns automatically
gender_col = find_column(master_df, 'gender', 'Gender', 'sex', 'Sex', 'customer_gender', 'gendor')
age_col = find_column(master_df, 'age', 'Age', 'customer_age', 'years', 'Age_Yrs')
marital_col = find_column(master_df, 'marital_status', 'Marital_Status', 'marital', 'Marital', 'marital status', 'martial')
occupation_col = find_column(master_df, 'occupation', 'Occupation', 'job', 'Job', 'profession', 'work')
age_group_col = find_column(master_df, 'age_group', 'Age_Group', 'agegroup', 'age range', 'agegroup')

# 1. Gender Distribution
plt.subplot(2, 3, 1)
if gender_col:
    gender_counts = master_df[gender_col].value_counts()
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
    plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', 
            colors=colors[:len(gender_counts)], startangle=90, 
            explode=[0.05]*len(gender_counts))
    plt.title('Gender Distribution', fontsize=14, fontweight='bold')
else:
    plt.text(0.5, 0.5, 'Gender Data Not Available', ha='center', va='center', fontsize=12)
    plt.title('Gender Distribution', fontsize=14, fontweight='bold')

# 2. Age Distribution
plt.subplot(2, 3, 2)
if age_col:
    age_data = master_df[age_col].dropna()
    plt.hist(age_data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(age_data.mean(), color='red', linestyle='dashed', linewidth=2, 
                label=f'Mean: {age_data.mean():.1f}')
    plt.axvline(age_data.median(), color='green', linestyle='dashed', linewidth=2,
                label=f'Median: {age_data.median():.1f}')
    plt.xlabel('Age', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title('Age Distribution', fontsize=14, fontweight='bold')
    plt.legend()
    plt.grid(True, alpha=0.3)
else:
    plt.text(0.5, 0.5, 'Age Data Not Available', ha='center', va='center', fontsize=12)
    plt.title('Age Distribution', fontsize=14, fontweight='bold')

# 3. Marital Status
plt.subplot(2, 3, 3)
if marital_col:
    marital_counts = master_df[marital_col].value_counts()
    bars = plt.bar(range(len(marital_counts)), marital_counts.values, color='lightcoral', edgecolor='black')
    plt.xlabel('Marital Status', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.title('Marital Status Distribution', fontsize=14, fontweight='bold')
    plt.xticks(range(len(marital_counts)), marital_counts.index, rotation=45)
    for i, v in enumerate(marital_counts.values):
        plt.text(i, v + 5, str(v), ha='center', fontweight='bold')
else:
    plt.text(0.5, 0.5, 'Marital Status Data Not Available', ha='center', va='center', fontsize=12)
    plt.title('Marital Status Distribution', fontsize=14, fontweight='bold')

# 4. Occupation (Top 10)
plt.subplot(2, 3, 4)
if occupation_col:
    occupation_counts = master_df[occupation_col].value_counts().head(10)
    plt.barh(occupation_counts.index, occupation_counts.values, color='lightgreen', edgecolor='black')
    plt.xlabel('Count', fontsize=12)
    plt.ylabel('Occupation', fontsize=12)
    plt.title('Top 10 Occupations', fontsize=14, fontweight='bold')
    plt.gca().invert_yaxis()
else:
    plt.text(0.5, 0.5, 'Occupation Data Not Available', ha='center', va='center', fontsize=12)
    plt.title('Top 10 Occupations', fontsize=14, fontweight='bold')

# 5. Age by Gender (Boxplot) - FIXED to remove warning
plt.subplot(2, 3, 5)
if gender_col and age_col:
    # Fixed version without the deprecation warning
    sns.boxplot(x=gender_col, y=age_col, hue=gender_col, data=master_df, 
                palette='Set2', legend=False)
    plt.xlabel('Gender', fontsize=12)
    plt.ylabel('Age', fontsize=12)
    plt.title('Age Distribution by Gender', fontsize=14, fontweight='bold')
else:
    plt.text(0.5, 0.5, 'Age or Gender Data Not Available', ha='center', va='center', fontsize=12)
    plt.title('Age Distribution by Gender', fontsize=14, fontweight='bold')

# 6. Age Group Distribution
plt.subplot(2, 3, 6)
if age_group_col:
    age_group_counts = master_df[age_group_col].value_counts().sort_index()
    plt.bar(range(len(age_group_counts)), age_group_counts.values, color='orange', edgecolor='black')
    plt.xlabel('Age Group', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.title('Age Group Distribution', fontsize=14, fontweight='bold')
    plt.xticks(range(len(age_group_counts)), age_group_counts.index, rotation=45)
elif age_col:
    # Create age groups automatically if age column exists
    bins = [0, 18, 25, 35, 45, 55, 65, 100]
    labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '65+']
    age_groups = pd.cut(master_df[age_col], bins=bins, labels=labels)
    age_group_counts = age_groups.value_counts().sort_index()
    plt.bar(range(len(age_group_counts)), age_group_counts.values, color='orange', edgecolor='black')
    plt.xlabel('Age Group', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.title('Age Group Distribution (Auto-created)', fontsize=14, fontweight='bold')
    plt.xticks(range(len(age_group_counts)), age_group_counts.index, rotation=45)
else:
    plt.text(0.5, 0.5, 'Age Group Data Not Available', ha='center', va='center', fontsize=12)
    plt.title('Age Group Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Print summary of what was plotted
print("\n" + "="*60)
print("VISUALIZATION SUMMARY")
print("="*60)
print(f"✓ Gender plot: {'Available' if gender_col else 'Not available'}")
print(f"✓ Age plot: {'Available' if age_col else 'Not available'}")
print(f"✓ Marital Status plot: {'Available' if marital_col else 'Not available'}")
print(f"✓ Occupation plot: {'Available' if occupation_col else 'Not available'}")
print(f"✓ Age by Gender plot: {'Available' if (gender_col and age_col) else 'Not available'}")
print(f"✓ Age Group plot: {'Available' if (age_group_col or age_col) else 'Not available'}")
```

```python
# Sales and Revenue Analysis
try:
    # Try to load from CSV (adjust filename as needed)
    orders = pd.read_csv('orders.csv')
    print("Orders data loaded successfully")
except FileNotFoundError:
    print("Orders file not found. Creating sample data...")
    # Create sample data if file doesn't exist
    np.random.seed(42)
    dates = pd.date_range('2023-01-01', '2023-12-31', freq='D')
    orders = pd.DataFrame({
        'order_date': np.random.choice(dates, 1000),
        'sales_quantity': np.random.randint(1, 10, 1000),
        'sales_amount': np.random.uniform(10, 500, 1000)
    })
    orders['sales_amount'] = orders['sales_amount'].round(2)

# Convert order_date to datetime
orders['order_date'] = pd.to_datetime(orders['order_date'])

# Create revenue column if not exists
if 'sales_amount' in orders.columns:
    orders['revenue'] = orders['sales_amount']
else:
    orders['revenue'] = orders['sales_quantity'] * orders['sales_amount']

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Daily Orders Trend
daily_orders = orders.set_index('order_date').resample('D').size()
axes[0,0].plot(daily_orders.index, daily_orders.values, color='blue', linewidth=2)
axes[0,0].set_title('Daily Orders Trend', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Date')
axes[0,0].set_ylabel('Number of Orders')
axes[0,0].tick_params(axis='x', rotation=45)

# Monthly Revenue Trend
monthly_revenue = orders.set_index('order_date').resample('M')['revenue'].sum()
axes[0,1].plot(monthly_revenue.index, monthly_revenue.values, color='green', marker='o', linewidth=2)
axes[0,1].set_title('Monthly Revenue Trend', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Month')
axes[0,1].set_ylabel('Revenue')
axes[0,1].tick_params(axis='x', rotation=45)

# Revenue Distribution
axes[1,0].hist(orders['revenue'].dropna(), bins=50, color='orange', edgecolor='black', alpha=0.7)
axes[1,0].set_title('Revenue Distribution', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Revenue')
axes[1,0].set_ylabel('Frequency')

# Sales Quantity Distribution
if 'sales_quantity' in orders.columns:
    axes[1,1].hist(orders['sales_quantity'].dropna(), bins=30, color='red', edgecolor='black', alpha=0.7)
    axes[1,1].set_title('Sales Quantity Distribution', fontsize=14, fontweight='bold')
    axes[1,1].set_xlabel('Quantity')
    axes[1,1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
